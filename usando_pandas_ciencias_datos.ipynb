```python
import pandas as pd
df = pd.read_csv("jugadores_futbol .csv")
df
df.index = ["primero", "segundo", "tercero", "cuarto", "quinto"]
df
df = df.set_index("jugador")
df
df_ordenado = df.sort_values(by="goles", ascending=False)
df_ordenado
df["goles_por_partido"] = df["goles"] / df["partidos_jugados"]
df
df[df["goles"] > 25]
df.describe()

df.info()
df.isnull().sum()
df.columns = df.columns.str.strip().str.lower()
df
df = df[df["partidos_jugados"] > 0]
df["media_goles"] = df["goles"] / df["partidos_jugados"]
df
df["media_goles"] = (df["goles"] / df["partidos_jugados"]).round(2)
df
df["goles"].mean()
df





```
