```python

while(True):
    largo=float(input("ingrese el largo del paralepipedo:"))
    ancho=float(input("ingrese el ancho del paralepipedo:"))
    altura=float(input("ingrese la altura del paralepipedo:"))
    vol=largo*ancho*altura
    print("el volumen del paralelepipedo es",vol)
    option=input("Desea continuar: s/n")
    if option=='n': break









```
